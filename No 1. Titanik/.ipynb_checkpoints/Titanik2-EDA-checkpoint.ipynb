```python
import missingno as msno
import warnings
warnings.filterwarnings('ignore')


%matplotlib inline
```

```python
df_train = pd.read_csv('./Dataset/train.csv')
df_test = pd.read_csv('./Dataset/test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
for col in df_train.columns:
    print('The NaN value of column {:>10} is {:.2f}'.format(col,100 * df_train[col].isnull().sum() / df_train[col].shape[0]))
```

```python
msno.matrix(df = df_train.iloc[:,:], figsize=(8,8), color=(0.8, 0.5, 0.2))
```

```python
f, ax = plt.subplots(1, 2, figsize=(18,8))
df_train['Survived'].value_counts().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Dead for visualizing bar chart')
df_train['Survived'].value_counts().plot.pie(autopct='%.1f%%',ax=ax[1])
plt.show()
```

```python
#df_train[['Pclass', 'Survived']].groupby('Pclass', as_index=False).count()
df_train[['Pclass', 'Survived']].groupby('Pclass', as_index=True).count().plot.bar()
plt.show()
```

```python
pd.crosstab(df_train['Pclass'], df_train['Survived'], margins=True).style.background_gradient(cmap='summer_r')
```

```python
print('{:>10} is {:.3f} Years'.format('   ', df_train['Age'].max()))
print('{:>10} is {:.3f} Years'.format('   ', df_train['Age'].min()))
print('{:>10} is {:.3f} Years'.format(' ', df_train['Age'].mean()))
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,8))
sns.kdeplot(df_train[df_train['Survived'] == 1]['Age'], ax=ax)
sns.kdeplot(df_train['Age'][df_train['Survived'] == 0], ax=ax)
plt.legend(['Survived', 'Dead'])
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
df_train['Age'][df_train['Pclass'] == 1].plot.kde()
df_train['Age'][df_train['Pclass'] == 2].plot.kde()
df_train['Age'][df_train['Pclass'] == 3].plot.kde()
plt.title('Age distribution depends on the Pclasses')
plt.legend(['First', 'Second', 'Third'])
plt.show()

```

```python
f, ax = plt.subplots(1,2, figsize=(18,8))
sns.violinplot("Pclass", 'Age', hue='Survived', data=df_train, scale='count', split=True, ax=ax[0])
```

```python
f, ax = plt.subplots(1,1,figsize=(8,8))
df_train[['Embarked', 'Survived']].groupby(['Embarked'], as_index=True).mean().sort_values(by='Survived', ascending=False).plot.bar(ax=ax)
```

```python
f, ax = plt.subplots(2,2, figsize=(20,15))
sns.countplot('Embarked', data=df_train, ax=ax[0,0])
ax[0,0].set_title('(1) No. Of Passengers Boarded')
sns.countplot('Embarked', hue='Sex', data=df_train, ax=ax[0,1])
ax[0,1].set_title('(2) Embarked graph per Sex')
sns.countplot('Embarked', hue='Pclass', data=df_train, ax=ax[1,0])
ax[1,0].set_title('(3) Embarked graph per Pclass')
sns.countplot("Embarked", hue='Survived', data=df_train, ax=ax[1,1])
ax[1,1].set_title('(4) Embarked graph per Survived')
```

```python
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch'] + 1
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch'] + 1
```

```python
print('The maximum size is {:.1f}'.format(df_train['FamilySize'].max()))
print("The minimum size if {:.1f}".format(df_train['FamilySize'].min()))
print("The mean size if {:.1f}".format(df_train['FamilySize'].mean()))
```

```python
f, ax = plt.subplots(1,3, figsize=(40,10))
sns.countplot('FamilySize', data=df_train, ax=ax[0])
ax[0].set_title('FamilySize numbers')
sns.countplot('FamilySize', hue='Survived', data=df_train, ax=ax[1])
ax[1].set_title('The survived families')
df_train[['Survived', 'FamilySize']].groupby(['FamilySize'], as_index=True).mean().sort_values(by='Survived', ascending=False).plot.bar(ax=ax[2])
ax[2].set_title('The survived mean of families numbers')
plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))
g = sns.distplot(df_train['Fare'], color='b', label='Skewness : {:.2f}'.format(df_train['Fare'].skew()), ax=ax)
g = g.legend(loc='best')
```

```python
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean()

df_train['Fare'] = df_train['Fare'].map(lambda i: np.log(i) if i>0 else 0)
df_test['Fare'] = df_test['Fare'].map(lambda i: np.log(i) if i>0 else 0)
```

```python
fig, ax = plt.subplots(1,1, figsize = (8,8))
g = sns.distplot(df_train['Fare'], color='b', label = 'Skewness : {:.2f}'.format(df_train['Fare'].skew()), ax=ax)
g = g.legend(loc='best')
```
